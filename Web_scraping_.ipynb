```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://www.imdb.com/chart/top/"
```

```python
headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept-Language":"en-US,en;q=0.9"
}
```

```python
response=requests.get(url, headers=headers)
```

```python
print("status:",response.status_code)
```

```python
soup=BeautifulSoup(response.text,"html.parser")
```

```python
movies=soup.find_all("li",class_="ipc-metadata-list-summary-item")
print(f"Movie Count :{len(movies)}")
```

```python
data=[]
for m in movies:
  title_t=m.find("h3", class_="ipc-title__text")
  title=title_t.text.strip() if title_t else "NA"

  rank_t=m.find("div", class_="ipc-signpost__text")
  rank=rank_t.text.split("#")[1] if rank_t else "NA"

  year_t=m.find("span", class_="sc-a55f6282-6 iMumIM cli-title-metadata-item")
  year=year_t.text.strip() if year_t else "NA"

  rating_t=m.find("span", class_="ipc-rating-star--rating")
  rating=rating_t.text.strip() if rating_t else "NA"

  data.append(
      {
        "Title":title,
        "Rank":rank,
        "Year":year,
        "Rating":rating
      })
df=pd.DataFrame(data)
print(df.head())
```

```python
df.to_csv("Web_scraped_IMDB_file.csv",index=False)
print("Saved Successfully")
```
